```python
import pandas as pd
import datetime as dt
```

```python
df_el_nino = pd.read_csv('Datasets\el_nino_la_nina_1999_2019.csv')
display(df_el_nino)
print(df_el_nino['start year'])
```

```python
print(pd.to_datetime(df_el_nino['start year'], format='%Y/01/01').dt.year)
```

```python
anos_desejados = range(df_el_nino['start year'].min(), df_el_nino['start year'].max() + 1)

# Reindexe o DataFrame com os anos desejados e preencha os valores ausentes na coluna Severidade com "Não houve"
df_el_nino = df_el_nino.set_index('start year').reindex(anos_desejados).fillna('Não houve').reset_index()

# Renomeie a coluna Ano de volta
df_el_nino = df_el_nino.rename(columns={'index': 'Ano'})

print(df_el_nino)
```

```python
df_temp = df_el_nino.groupby(['start year', 'severity']).agg({"phenomenon":"nunique"}).reset_index()
df_temp
```

```python
df_temp = df_temp.sort_values(by='start year', ascending=False)
df_temp
```

```python
df_temp = pd.to_datetime(df_temp['start year'], format='%Y')
print(df_temp)
```
